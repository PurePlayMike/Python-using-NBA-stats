# Converting stats to fantasy points

```python
import pandas as pd

player_proj = pd.read_csv('../NBA Projection/total_stats_df.csv')
player_info = pd.read_csv('../NBA Projection/player_info_df.csv')
```

```python
df_proj_stats = player_proj.loc[:, 
               ['player_id', 'player_name', 'proj_season_id', 'proj_pts','proj_min','proj_fgm','proj_fga','proj_fg3m','proj_fg3a','proj_ftm','proj_fta',
                'proj_oreb','proj_dreb','proj_ast','proj_stl','proj_tov', 'proj_blk']]
```

```python
df_proj_stats['proj_reb'] = df_proj_stats['proj_oreb'] + df_proj_stats['proj_dreb']
df_proj_stats['proj_fg%'] = ((df_proj_stats['proj_fgm'] / df_proj_stats['proj_fga'])*100)
df_proj_stats['proj_ft%'] = ((df_proj_stats['proj_ftm'] / df_proj_stats['proj_fta'])*100)
df_proj_stats['proj_3p%'] = ((df_proj_stats['proj_fg3m'] / df_proj_stats['proj_fg3a'])*100)
df_proj_stats['proj_ppm'] = df_proj_stats['proj_pts'] / df_proj_stats['proj_min']
```

```python
df_proj_stats.rename(columns = {'proj_season_id':'season_id',
                               'proj_pts': 'pts',
                                'proj_min':'min',
                                'proj_ppm':'ppm',
                                'proj_fgm':'fgm',
                                'proj_fga':'fga',
                                'proj_fg%':'fg%',
                                'proj_fg3m':'fg3m',
                                'proj_fg3a':'fg3a',
                                'proj_3p%':'3p%',
                                'proj_ftm':'ftm',
                                'proj_fta':'fta',
                                'proj_ft%':'ft%',
                                'proj_reb':'reb',
                                'proj_ast':'ast',
                                'proj_stl':'stl',
                                'proj_tov':'tov',
                               'proj_blk':'blk'}, inplace = True)
```

```python
df_proj_stats.sample(5)
```

```python
df_proj_stats = df_proj_stats.reindex(['player_id', 'player_name', 'season_id', 'min', 'pts', 'ppm',
                                       'fg%', '3p%', 'ft%', 'reb', 'ast', 'stl', 'tov', 'blk'], axis=1)
df_proj_stats
```

```python
def fantasy_pts_converter(row):
    pts = row['pts']
    reb = row['reb']
    ast = row['ast']
    stl = row['stl']
    tov = row['tov']
    blk = row['blk'] 
    fantasy_pts = pts + (1.25*reb) + (1.5*ast) + (2*stl) + (2*blk) - (.5*tov) 
    return fantasy_pts
```

```python
fantasy_value = df_proj_stats.apply(fantasy_pts_converter, axis = 1)
```

```python
df_proj_stats['fantasy_pts'] = fantasy_value
```

```python
df_proj_stats['fppm'] = df_proj_stats['fantasy_pts'] / df_proj_stats['min']
```

```python
df_proj_stats.sort_values('fantasy_pts', ascending = False)
```

```python
df_for_export = round(df_proj_stats.sort_values('fantasy_pts', ascending = False),2)
```

```python
df_for_export.to_csv('../NBA Projection/projected_DK_fantasy_pts_final_2018_19.csv', index=False)
```
