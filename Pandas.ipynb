```python
# round(ft_prec,1); round ft_perc to 1 decimal place. using + .00001 to make sure dont divide by 0

def ft_calculator(row):
    ft_perc = row['ftm']/(row['fta'] + .00001)
    return round(ft_perc,2)
```

```python
df['ft_perc'] = df.apply(ft_calculator, axis=1)
```

```python
df.sample(5)
```

```python
#Career FT %... 1)group by player id/name 2)sum 3)apply ft_calc function
career = df.groupby('player_name').sum().apply(ft_calculator, axis=1)
```
